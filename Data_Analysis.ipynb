```python
import pandas as pd
df1=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv') #Ball by ball data
df2=pd.read_csv('IPL_Matches_2008_2022.csv') # Every match data
```

**Joining the two datasets on common feature 'ID'**

```python
merged_df = pd.merge(df1, df2, on='ID')
merged_df
```

```python
merged_df.columns
```

**1. Title Winner Each year**

```python
import pandas as pd

# Read the merged CSV file into a DataFrame
df = merged_df

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group the DataFrame by 'Season'
season_groups = df.groupby('Season')

# Create an empty list to store the extracted rows
extracted_rows = []

# Iterate over each season group
for season, group in season_groups:
    # Find the index of the row with the maximum 'Date' value
    max_date_idx = group['Date'].idxmax()
    
    # Use the index to extract the row with the maximum 'Date' value
    max_date_row = group.loc[max_date_idx]
    
    # Append the extracted row to the list
    extracted_rows.append(max_date_row)
    
# Create a new DataFrame from the extracted rows
winning_teams = pd.DataFrame(extracted_rows, columns=['Season', 'WinningTeam'])

# Reset the index to show numbers starting from 1
winning_teams = winning_teams.reset_index(drop=True)
winning_teams.index += 1

# Print the new DataFrame
winning_teams
```

**2. Orange Caps (Batsman who scored most runs each season)**

```python
# Group the data by season and batsman and sum the runs scored
season_batsman_runs = merged_df.groupby(['Season', 'batter'])['batsman_run'].sum()

# Reset the index to turn the grouped columns back into regular columns
season_batsman_runs = season_batsman_runs.reset_index()

# Find the index of the maximum runs scored by each season
max_runs_season_batsman = season_batsman_runs.groupby('Season')['batsman_run'].idxmax()

# Use the index to get the rows with the maximum runs scored by each season and batsman
max_runs_season_batsman_rows = season_batsman_runs.loc[max_runs_season_batsman]

# Print the names of the players who scored maximum runs each season
for season, batter in zip(max_runs_season_batsman_rows['Season'], max_runs_season_batsman_rows['batter']):
    print(f"{season}            {batter} ")
```

**3. Purple cap (Bowlers who got most wickets each season)**

```python
# Group the data by season and bowler and count the number of wickets taken
season_bowler_wickets = merged_df.groupby(['Season', 'bowler'])['isWicketDelivery'].sum()

# Reset the index to turn the grouped columns back into regular columns
season_bowler_wickets = season_bowler_wickets.reset_index()

# Find the index of the maximum wickets taken by each season
max_wickets_season_bowler = season_bowler_wickets.groupby('Season')['isWicketDelivery'].idxmax()

# Use the index to get the rows with the maximum wickets taken by each season and bowler
max_wickets_season_bowler_rows = season_bowler_wickets.loc[max_wickets_season_bowler]

# Print the names of the bowlers who got maximum wickets each season
for season, bowler in zip(max_wickets_season_bowler_rows['Season'], max_wickets_season_bowler_rows['bowler']):
    print(f"{season}       {bowler} ")
```

**4. Tournament 4's per season**

```python
import pandas as pd

df=merged_df# Filter the data to only include rows where the batsman scored 4 runs
fours = df[df['batsman_run'] == 4]

# Group the filtered data by season and count the number of rows in each group
fours_by_season = fours.groupby('Season')['batsman_run'].count()

# Convert the Pandas GroupBy object to a Series object
fours_by_season = pd.Series(fours_by_season)

# Print the resulting Pandas Series object
fours_by_season
```

**5. Tournament 6's per season**

```python
# Filter the rows where 6 runs were scored
sixes = merged_df[merged_df['total_run'] == 6]

# Group the data by season and count the number of 6's
season_sixes = sixes.groupby('Season')['total_run'].count()

# Return the Pandas Series object
season_sixes
```
